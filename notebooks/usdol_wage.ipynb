```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/usdol_wage.csv')
df
```

```python
df.value_counts('naics_code_description')
```

```python
df[['naic_cd','naics_code_description']].info()
```

```python
naic = df[['naic_cd','naics_code_description']].dropna().drop_duplicates()
naic['construct'] = naic['naics_code_description'].str.lower().str.contains('construct')
naic.to_csv('checknaics.csv')
```

```python
df = df.merge(naic[naic['construct']], on = ['naic_cd','naics_code_description'], how = 'inner')
```

```python
df.to_csv('../data/processed/usdol_wage_construction.csv', index = False)
```
