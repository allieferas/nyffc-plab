```python
import pandas as pd

# data
nyc_awarded_contracts_path = "../data/raw/NYC_Awarded_Contracts.csv"

df_contracts = pd.read_csv(nyc_awarded_contracts_path)


```

```python
print(f"Total records in NYC Awarded Contracts: {len(df_contracts)}")
```

```python

#missing values
print("\nMissing Values:\n", df_contracts.isnull().sum()[df_contracts.isnull().sum() > 0])

#data types
print("\nData Types:\n", df_contracts.dtypes)


df_contracts.head()
```

```python
#money-related columns
money_cols = [
    "Prime Contract Current Amount", "Prime Contract Original Amount",
    "Prime Vendor Spend to Date", "Sub Contract Current Amount",
    "Sub Contract Original Amount", "Sub Vendor Paid to Date"
]

# Clean and convert monetary columns
for col in money_cols:
    df_contracts[col] = (
        df_contracts[col]
        .astype(str)  
        .str.replace("[\$,]", "", regex=True) 
        .str.strip()  
        .replace(["", " ", "-", "NaN", "nan"], "0")  
        .astype(float)  
    )

print("Monetary values cleaned and converted to numeric format.")
```

```python
prime_cols = [c for c in df_contracts.columns if 'prime' in c.lower()]
sub_cols = [c for c in df_contracts.columns if 'sub' in c.lower()]
```

```python
nyc = df_contracts
nyc.loc[nyc['Vendor Record Type']=='Prime Vendor','NAME1'] = nyc.loc[nyc['Vendor Record Type']=='Prime Vendor','Prime Vendor']
```

```python
nyc.loc[nyc['Vendor Record Type']=='Sub Vendor','NAME1'] = nyc.loc[nyc['Vendor Record Type']=='Sub Vendor','Sub Vendor']
```

```python
df_contracts
```

```python
# Standardize "Contract Includes Sub Vendors" column
df_contracts["Contract Includes Sub Vendors"] = df_contracts["Contract Includes Sub Vendors"].str.upper().str.strip()

# Fill missing values with "No" if the column is NOT "YES"
df_contracts.loc[df_contracts["Contract Includes Sub Vendors"] != "YES", "Contract Includes Sub Vendors"] = "No"

print(df_contracts["Contract Includes Sub Vendors"].value_counts())
```

```python
#date columns
date_cols = [
    "Prime Contract Start Date", "Prime Contract End Date",
    "Sub Contract Start Date", "Sub Contract End Date",
    "Prime Contract Registration Date"
]

# Convert date columns to datetime
for col in date_cols:
    df_contracts[col] = pd.to_datetime(df_contracts[col], errors="coerce")

# Check for any null values in date columns after conversion
print("Date columns converted.\nMissing Dates:\n", df_contracts[date_cols].isnull().sum())

##need to determine if i should keep as datetime or leave missing values
```

```python
# Define the specific subcontract columns to fill
subcontract_fill_cols = [
    "Sub Vendor M/WBE Category", "Sub Contract Purpose",
    "Subcontract Status", "Sub Contract Industry"
]

# Define condition: When "Sub Vendor" is blank or "-"
condition = df_contracts["Sub Vendor"].isna() | (df_contracts["Sub Vendor"].str.strip() == "-")

# Fill the specified columns with "Not Applicable" only where "Sub Vendor" is blank
df_contracts.loc[condition, subcontract_fill_cols] = "Not Applicable"

```

```python
df_contracts.head()
```

```python
# Standardize "Contract Includes Sub Vendors" column
df_contracts["Contract Includes Sub Vendors"] = df_contracts["Contract Includes Sub Vendors"].str.strip().str.lower()

# Define condition: "Contract Includes Sub Vendors" is "no" but "Sub Vendor" is NOT blank or "-"
condition = (df_contracts["Contract Includes Sub Vendors"] == "no") & \
            (~df_contracts["Sub Vendor"].isna()) & \
            (df_contracts["Sub Vendor"].str.strip() != "") & \
            (df_contracts["Sub Vendor"].str.strip() != "-")  # Ignore "-"

# Filter and display the affected rows
inconsistent_rows = df_contracts[condition]

# Show result
if inconsistent_rows.empty:
    print(" No inconsistencies found: All 'No' entries in 'Contract Includes Sub Vendors' have blank 'Sub Vendor' fields.")
else:
    print("Inconsistencies found! These rows have 'Sub Vendor' filled despite 'Contract Includes Sub Vendors' being 'No':")
    print(inconsistent_rows[["Contract Includes Sub Vendors", "Sub Vendor"]].head(10))  # Show first 10 inconsistent rows
```

```python
# Replace only "-" values in "Sub Vendor" with "Not Applicable"
df_contracts["Sub Vendor"] = df_contracts["Sub Vendor"].replace("-", "Not Applicable")

print("'Sub Vendor' values updated: '-' replaced with 'Not Applicable'.")
```

```python
# Replace only "-" values in "Sub Contract Reference ID" with "Not Applicable"
df_contracts["Sub Contract Reference ID"] = df_contracts["Sub Contract Reference ID"].replace("-", "Not Applicable")


print(" 'Sub Contract Reference ID' values updated: '-' replaced with 'Not Applicable'.")
```

```python
# Define columns to update
business_cols = ["Sub Woman Owned Business", "Sub Emerging Business"]

# Define condition: "Contract Includes Sub Vendors" is "No"
condition = df_contracts["Contract Includes Sub Vendors"].str.strip().str.lower() == "no"

# Fill selected columns with "Not Applicable" where condition is met
df_contracts.loc[condition, business_cols] = "Not Applicable"
```

```python
# Define percentage columns to clean
percentage_cols = ["% COVID Spending", "% Other Spending"]

# Replace blanks, NaN, and "-" with 0
df_contracts[percentage_cols] = df_contracts[percentage_cols].replace(["", "-", None], "0").fillna("0")

# Convert to numeric (if needed)
df_contracts[percentage_cols] = df_contracts[percentage_cols].astype(float)
```
